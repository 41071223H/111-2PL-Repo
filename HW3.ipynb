```python
import requests
import json
import time
from time import sleep
import pandas as pd
```

```python
from bs4 import BeautifulSoup
htmlRUL='https://www.ptt.cc/bbs/NTNUITE/index'
htmlRUL
```

```python
pages = range(1,10)
allurl = list()
urlfirst = 'https://www.ptt.cc/'
for page in pages:
    link = htmlRUL + str(page) + '.html'
    response = requests.get(link)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text,'html.parser')
    titles = soup.find_all(class_='title')
    print(titles)
    url = soup.find_all('a')
    for a in url:
        href = a.get('href')
        if href and href.startswith('/'):
            temp = urlfirst + href
            allurl.append(temp)
```

```python
#找出標題
for title in titles:
        link_tag = title.find('a')
        if link_tag:
            text = link_tag.text
            allurl.append(text)
            
```
